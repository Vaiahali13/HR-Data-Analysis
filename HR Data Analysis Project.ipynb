### 2)How does attrition vary by department?

```python
attr_by_department = df.groupby('Department')['Attrition'].value_counts().unstack()
print("Attrition by Department:", attr_by_department)
plt.figure(figsize=(10, 6))
plt.title('Attrition by Department')
plt.bar(attr_by_department.index, attr_by_department['Yes'], width=0.4, label='Attrition: Yes')
plt.bar(attr_by_department.index, attr_by_department['No'], width=0.4, label='Attrition: No', bottom=attr_by_department['Yes'])
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.show()
```

there are 3 no.of department are there 1)sales 2) Research and development 3)HR department
"research and development" department have more number of attrition as compared to other two department,research and development in there more number of employees,this department affecting the most for attrition
"Human Resource" Department have least attrition

### 10)Attrition Rate by Age Group

```python
# Define custom age groups
def get_age_group(age):
    if age <= 24:
        return '18-24'
    elif age <= 34:
        return '25-34'
    elif age <= 44:
        return '35-44'
    elif age <= 54:
        return '45-54'
    elif age <= 64:
        return '55-64'
    else:
        return '65+'

# Apply the function to create the 'AgeGroup' column
df['AgeGroup'] = df['Age'].apply(get_age_group)

# Group data by age group and attrition, and count occurrences
attrition_by_agegroup = df.groupby(['AgeGroup', 'Attrition']).size().unstack()
print(attrition_by_agegroup)

xpos = np.arange(len(attrition_by_agegroup.index))
color_yes = 'skyblue'
color_no = 'green'

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(xpos-0.2, attrition_by_agegroup['Yes'], width=0.4, label='Attrition: Yes',color=color_yes)
plt.bar(xpos+0.2, attrition_by_agegroup['No'], width=0.4, label='Attrition: No',color=color_no)
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
# Set x-ticks and labels
plt.xticks(xpos, attrition_by_agegroup.index)
plt.xticks(rotation=60)
plt.legend(title='Attrition')
plt.show()

```

employees in age of 25 to 34 are more likely to leave the job
